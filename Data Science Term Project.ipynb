```python

%matplotlib inline
import matplotlib.pyplot as mp
import urllib2
import urllib
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import re

#variables
topSongs = [];

def getTop100(page_url):
    user_agent = {'User-agent': 'Mozilla/5.0'}
    page_detail = requests.get(page_url, headers=user_agent);
    return page_detail;

top100 = getTop100("http://www.metrolyrics.com/top100.html")
soup = BeautifulSoup(top100.text,'lxml');
top20 = soup.body.find_all("ul", {"class": "top20"})[0].find_all("li");
top60 = soup.body.find_all("div", {"class": "song-list"})[0]
print top60.find_all("ul", {"class":"song-list"});
for listItem in top20:
    linkToSong = listItem.find_all("span")[1].find_all("a")[0];
    endIndex = re.search(r'\b(onmousedown)\b', str(linkToSong)).start();
    topSongs.append(str(linkToSong)[9:endIndex-2]) #link to specific song
  
```
